```python
import re

with open('row.txt', 'r', encoding='utf-8') as file:
    content = file.read()

pattern = re.compile(
    r'^\d+\.\n'  # Номер позиции
    r'(.+)\n'    # Наименование товара
    r'(\d+,\d{3}) x (\d+,\d{2})\n'  # Количество и цена за единицу
    r'(\d+,\d{2})$',  
    re.MULTILINE
)

matches = pattern.findall(content)

products = []
for match in matches:
    name = match[0].strip()
    quantity = float(match[1].replace(',', '').replace('.', ''))
    unit_price = float(match[2].replace(',', '').replace('.', ''))
    total_price = float(match[3].replace(',', '').replace('.', ''))
    products.append({
        'name': name,
        'quantity': quantity,
        'unit_price': unit_price,
        'total_price': total_price
    })

for product in products:
    print(f"Товар: {product['name']}")
    print(f"Количество: {product['quantity']}")
    print(f"Цена за единицу: {product['unit_price']}")
    print(f"Общая стоимость: {product['total_price']}")
    print('---')
```
